```python
from articlehelper import ArticleHandler, RePostprocessor, map_category_ukrpravda, map_category_bruk
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import logging
MASK = 'data\\news'
zip_fname = MASK + '.zip'
train_fname = MASK + '_train.txt'
test_fname = MASK + '_test.txt'
```

```python
articles = ArticleHandler.read_zip(zip_fname)
```

```python
for doc in articles:
    try:
        filename = doc['filename']
        doc['category'] = map_category_ukrpravda(doc)
    except:
        logging.exception('Error with ' + filename)
df = pd.DataFrame.from_dict(articles)
df
```

```python
df_uk = df.loc[df['language'] == 'uk']
# categories = ['news', 'blog', 'columns']
# other_category = 'article'

# df_uk = df_uk.join(df_uk['autocategory'].map(lambda x: x if x in categories else other_category).rename('category'))
df_uk = df_uk.join(df_uk['text'].map(lambda x: len(x)).rename('length'))
df_uk = df_uk.join(df_uk['body'].map(lambda x: len(x)).rename('par_count'))
df_uk.groupby(['category']).count()
```

```python
df_uk.groupby(['category'])["length"].sum()
```

```python
df_uk.groupby(['category'])["par_count"].sum()
```

```python
categories = {'articles':350, 'columns':450, 'news':1600, 'blog':600}
print(len(df_uk))
df_sample = pd.DataFrame(columns=df_uk.columns)
print(len(df_sample))
for c, v in categories.items():
    dfc = df_uk[df_uk['category'] == c]
    print(len(dfc))
    df_sample = df_sample.append(dfc.sample(v))
print(len(df_sample))
print(df_sample.groupby(['category'])["par_count"].sum())
print(df_sample.groupby(['category'])["length"].sum())
print(df_sample.groupby(['category']).count())
```

```python
X = df_sample['text'].values
y = df_sample['category'].values
z = df_sample['filename'].values
X_train, X_test, y_train, y_test, z_train, z_test = train_test_split(X, y, z,  random_state=42, stratify = y, shuffle=True)
print(len(X_train), len(X_test))
```

### Категорії текстів у БрУК (відсотки визначено колективно 30.08.2014)

| Ідентифікатор | Категорія | Частка в корпусі | Опис текстів |
| ------------- |-----------|------------------|--------------|
| А | Преса | 25% | Репортажі, огляди, редакційні статті, листи до редакції; національні й регіональні видання; тематично - політика, спорт, суспільство, економіка й фінанси, короткі новини, культура - театр, література, музика, танці |
| B | Релігійна література | 3% | Книжки, періодика, брошури |
| С | Професійно-популярна література | 7% | Книжки й періодика; домоводство, ремесла, «сад і город», хобі, ремонт і будівництво, конструювання, музика й танці, домашні тварини, спорт, їжа й вино, подорожі, фермерство, робочі професії тощо |
| D | «Естетичні інформативні» тексти | 7% | Інформативні тексти, що не потрапляють в інші категорії, зокрема, біографії, мемуари, есеї, передмови, особисті листи, художня й мистецтвознавча критика, рекламні тексти |
| E | Адміністративні документи | 3% | Закони, урядові акти, звіти організацій/фондів/компаній, офіційні листи |
| F | Науково-популярна література | 5% | |
| G | Наукова література | 10% | Книжки й періодика; природничі й гуманітарні науки, техніка й інженерна справа |
| H | Навчальна література | 15% | Підручники, посібники, гуманітарні та природничі науки тощо |

Загалом тексти поділяються на два види – інформативні (призначені поінформувати читача) і художні (описують вигаданих персонажів і події). Інформативний вид охоплює категорії А-H, художній – лише I.

## Розділяємо корпус на 3 категорії: преса (A), наукова та професійна література (C, D, F, G, H), художня література (I)

```python
for x in articles:
    try:
        x['category'] = map_category_bruk(x)
    except:
        logging.exception(x['filename'])
df_bruk = pd.DataFrame.from_dict(articles)
df_bruk['category'].count()
```

```python
df_bruk.groupby(['category', 'id']).count()
```

```python
df_bruk = df_bruk.drop(df_bruk[df_bruk['category'] == 'other'].index)
df_bruk.groupby(['category']).count()
```

```python
categories = {'fiction':100, 'press':173, 'prof_science':77}
print(len(df_bruk))
df_sample = pd.DataFrame(columns=df_bruk.columns)
print(len(df_sample))
for c, v in categories.items():
    dfc = df_bruk[df_bruk['category'] == c]
    print(len(dfc))
    df_sample = df_sample.append(dfc.sample(v))
print(len(df_sample))
```

```python
X = df_sample['text'].values
y = df_sample['category'].values
z = df_sample['filename'].values
X_train, X_test, y_train, y_test, z_train, z_test = train_test_split(X, y, z,  random_state=42, stratify = y, shuffle=True)
print(len(X_train), len(X_test))
```
