```python
df_uk = df.loc[df['language'] == 'uk']
categories = ['news', 'blog', 'columns']
other_category = 'article'
df_uk = df_uk.join(df_uk['autocategory'].map(lambda x: x if x in categories else other_category).rename('category'))
df_uk = df_uk.join(df_uk['text'].map(lambda x: len(x)).rename('length'))
df_uk.groupby(['category']).count()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

X = df_uk['text'].values
y = df_uk['category'].values
z = df_uk['filename'].values
X_train, X_test, y_train, y_test, z_train, z_test = train_test_split(X, y, z,  random_state=42, #test_size=0.3, 
                                                                       stratify = y, shuffle=True)
print(len(X_train), len(X_test))
```

```python
import pymorphy2
from tokenize_uk import tokenize_uk

def tokenize(text):
    pars = tokenize_uk.tokenize_text(text)
    tokens = []
    for par in pars:
        for sent in par:
            tokens = tokens + sent
    return tokens

class Morphology():
    _morph = None
    def __init__(self, lang='uk'):
        Morphology.load(lang=lang)

    @staticmethod
    def load(path = None, lang='uk'):
        if (Morphology._morph == None):
            print('Morphology initialized for', lang)
            Morphology._morph = pymorphy2.MorphAnalyzer(lang=lang)

    def getAnalyzer(self):
        return Morphology._morph

def analyze_POS(tokens):
    morphy = Morphology()
    data = []
    for t in tokens:
        parsed = morphy.getAnalyzer().parse(t)
        if parsed:
            p = parsed[0]
            if p.tag.POS:
                d = p.tag.POS
            elif t[0].isdigit():
                d = 'NUMBER'
            else: d = p.normal_form
        else:
            d = 'UNDEF'
        data.append(d)
    return data
    
def POS_tokenize(text):
    return analyze_POS(tokenize(text))

s = 'Окрім того, зарплата першого заступника міністра економрозвитку та торгівлі Максима Нефьодова становила 92,3 тис. грн, заступників Юрія Бровченка - 92,4 тис. грн, Михайла Титарчука - 76,5 тис. грн, Наталії Микольської - 60,2 тис. грн, а держсекретаря Олексія Перевезенцева - 81,7 \xa0тис. грн.'
tokens = tokenize(s)
print(tokens)
morphy = Morphology()
print(morphy.getAnalyzer().parse(tokens[0]))
# print(analyze_POS_lem(tokens))
print(analyze_POS(tokens))    
```

```python
tfidf = TfidfVectorizer(tokenizer=POS_tokenize, ngram_range=(2, 4), min_df=10)
train = tfidf.fit_transform(X_train)
```

```python
test = tfidf.transform(X_test)
clf = MultinomialNB(alpha=0.05)
clf.fit(train.toarray(), y_train)
y_predicted = clf.predict(test.toarray())
```

```python
print(classification_report(y_test, y_predicted, clf.classes_))
```

```python
def top_features(vectorizer, clf, n):
    """Prints features with the highest coefficient values, per class"""
    feature_names = vectorizer.get_feature_names()
    for i, class_label in enumerate(clf.classes_):
        top = np.argsort(clf.coef_[i])
        reversed_top = top[::-1]
        print("%s:\n%s" % (class_label,
              "\n".join(feature_names[j] for j in reversed_top[:n])))
        
def show_most_informative_features(vectorizer, clf, i_label, n=20, joiner = '\n'):
    feature_names = vectorizer.get_feature_names()
    coefs_with_fns = sorted(zip(clf.coef_[i_label], feature_names), reverse=True)
    top = coefs_with_fns[:n]
    return joiner.join(['{}: {}'.format(x[1], x[0]) for x in top])
    
def show_all_top_features(vectorizer, clf, n=20, joiner = '\n'):
    if clf.classes_.shape[0] == clf.coef_.shape[0]:
        for i, class_label in enumerate(clf.classes_):
            print(class_label, ':\n', show_most_informative_features(vectorizer, clf, i, n, joiner), sep='')
    else:
        for i in range(clf.coef_.shape[0]):
            print(i, ':\n', show_most_informative_features(vectorizer, clf, i, n, joiner), sep='')        
```
